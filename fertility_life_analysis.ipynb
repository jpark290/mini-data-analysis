```python
# 1. Introduction
## This project explores the relationship between total fertility rate and life expectancy across countries.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Dataset Overview
"""- Source: `Country-data.csv`
   - 167 countries
   - No missing values"""

file_path = "Country-data.csv"
df = pd.read_csv(file_path)
```

```python
# 3. Summary Statistics
## [insert `describe()` output]

print(df.info())
print(df[['total_fer', 'life_expec']].describe())
```

```python
# 4. Correlation Analysis
## The correlation coefficient between fertility rate and life expectancy is -0.761, indicating a strong negative relationship.

correlation = np.corrcoef(df['total_fer'], df['life_expec'])[0,1]
print(f"Correlation coefficient between fertility rate and life expectancy: {correlation:.3f}")
```

```python
# 5. Visualization
## [Include scatter plot]

plt.scatter(df['total_fer'], df['life_expec'], alpha=0.7)
plt.title('Fertility Rate vs Life Expectancy')
plt.xlabel('Total Fertility Rate')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# 6. Conclusion
## Countries with higher fertility rates tend to have lower life expectancy, and vice versa.
### This may reflect development levels, access to healthcare, and socio-economic factors.
```
